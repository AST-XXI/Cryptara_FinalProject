```python
#Run total suggestions and compile into list
import warnings 
warnings.simplefilter(action='ignore')
import datetime;
ct = datetime.datetime.now()
print(f'Gemini scanner started at {ct}...')
from ipynb.fs.full.admin import *
from ipynb.fs.full.imported_file_gemini import *
```

```python
final_buy_daily['timeframe'] = 'DAILY'
final_sell_daily['timeframe'] = 'DAILY'
final_buy_weekly['timeframe'] = 'WEEKLY'
final_sell_weekly['timeframe'] = 'WEEKLY'
final_buy_fourh['timeframe'] = 'FOURHOUR'
final_sell_fourh['timeframe'] = 'FOURHOUR'
final_buy_twoh['timeframe'] = 'TWOHOUR'
final_sell_twoh['timeframe'] = 'TWOHOUR'
final_buy_oneh['timeframe'] = 'ONEHOUR'
final_sell_oneh['timeframe'] = 'ONEHOUR'
final_buy_thirtymin['timeframe'] = 'THIRTYMIN'
final_sell_thirtymin['timeframe'] = 'THIRTYMIN'
final_buy_fifteenmin['timeframe'] = 'FIFTEENMIN'
final_sell_fifteenmin['timeframe'] = 'FIFTEENMIN'
final_buy_fivemin['timeframe'] = 'FIVEMIN'
final_sell_fivenmin['timeframe'] = 'FIVEMIN'
final_buy_onemin['timeframe'] = 'ONEMIN'
final_sell_onemin['timeframe'] = 'ONEMIN'
```

```python
#BUY DATA, CONCAT BUY AND SELL DATA
total_buy = pd.concat([final_buy_weekly,final_buy_daily,final_buy_fourh,final_buy_twoh,final_buy_oneh,final_buy_thirtymin,final_buy_fifteenmin,final_buy_fivemin,final_buy_onemin], axis=0,join='inner')
total_buy.drop(columns=['BUY','SELL','NEUTRAL'], inplace=True)
total_buy['frequency_count'] = total_buy.index.value_counts()
total_buy.sort_values(['frequency_count','RECOMMENDATION'], ascending=False, inplace=True)
total_sell = pd.concat([final_sell_weekly,final_sell_daily,final_sell_fourh,final_sell_twoh,final_sell_oneh,final_sell_thirtymin,final_sell_fifteenmin,final_sell_fivenmin,final_sell_onemin], axis=0,join='inner')
total_sell.drop(columns=['BUY','SELL','NEUTRAL'], inplace=True)
total_sell['frequency_count'] = total_sell.index.value_counts()
total_sell.sort_values(['frequency_count','RECOMMENDATION'], ascending=False, inplace=True)
total_buy.to_csv('Data_files/output_files/total_buys_gemini.csv')
total_sell.to_csv('Data_files//output_files/total_sells_gemini.csv')
```

```python
#BUY DATA, BINNING_Categorize strong, moderate, light buys based on number of buy indicators
buy_bins=[0, 4, 6, 15]
buy_volume=['light','moderate','heavy']
total_buy["intensity"] = pd.cut(total_buy["frequency_count"], buy_bins, labels=buy_volume)
strong_buy = total_buy.loc[total_buy['intensity'] == 'heavy'].copy()
strong_buy = strong_buy[~strong_buy.index.duplicated(keep='last')]
strong_buy.to_csv('Data_files/output_files/strong_buy_gemini.csv', columns=[], header=False)
moderate_buy = total_buy.loc[total_buy['intensity'] == 'moderate'].copy()
moderate_buy = moderate_buy[~moderate_buy.index.duplicated(keep='last')]
moderate_buy.to_csv('Data_files/output_files/moderate_buy_gemini.csv', columns=[], header=False)
light_buy = total_buy.loc[total_buy['intensity'] == 'light'].copy()
light_buy = light_buy[~light_buy.index.duplicated(keep='last')]
light_buy.to_csv('Data_files/output_files/light_buy_gemini.csv', columns=[], header=False)
```

```python
#SELL DATA, BINNING_Categorize strong, moderate, light buys based on number of buy indicators
sell_bins=[0, 4, 6, 15]
sell_volume=['light','moderate','heavy']
total_sell["intensity"] = pd.cut(total_sell["frequency_count"], sell_bins, labels=sell_volume)
strong_sell = total_sell.loc[total_sell['intensity'] == 'heavy'].copy()
strong_sell = strong_sell[~strong_sell.index.duplicated(keep='last')]
strong_sell.to_csv('Data_files/output_files/strong_sell_gemini.csv', columns=[], header=False)
moderate_sell = total_sell.loc[total_sell['intensity'] == 'moderate'].copy()
moderate_sell = moderate_sell[~moderate_sell.index.duplicated(keep='last')]
moderate_sell.to_csv('Data_files/output_files/moderate_sell_gemini.csv', columns=[], header=False)
light_sell = total_sell.loc[total_sell['intensity'] == 'light'].copy()
light_sell = light_sell[~light_sell.index.duplicated(keep='last')]
light_sell.to_csv('Data_files/output_files/light_sell_gemini.csv', columns=[], header=False)
```

```python
print(f'Gemini scanner exported to CSV at {ct}')
```
