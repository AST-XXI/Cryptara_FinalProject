```python
#Extract all buy and sell recommendations from all analyses.
from ipynb.fs.full.admin import *
from tradingview_ta import TA_Handler, Interval, Exchange
ticker_data = Path('Data_files/total_crypto_list.csv')
ticker= pd.read_csv('Data_files/total_crypto_list.csv', header=None)
ticker= pd.DataFrame(ticker)
my_crypto_list = ticker[0].tolist()
screener="CRYPTO"
exchange="GEMINI"
df_daily = pd.DataFrame()
```

```python
#Obtain all buy and sell cryptos based on daily chart
staging_df = pd.DataFrame()
ticker_df_daily = pd.DataFrame()
for ticker in my_crypto_list:
    try:
        data = (TA_Handler(symbol=ticker,screener=screener,
                           exchange=exchange,interval=Interval.INTERVAL_1_DAY ).get_analysis().summary)
        symbol = ticker
        staging_df = list(data.values())
        final_df = (pd.DataFrame((data), index={ticker}))
        ticker_df_daily = ticker_df_daily.append(final_df)
    except:
        pass
    
final_buy_daily = ticker_df_daily.loc[ticker_df_daily['RECOMMENDATION'].str.contains('BUY|STRONG_BUY',regex=True)]
final_sell_daily = ticker_df_daily.loc[ticker_df_daily['RECOMMENDATION'].str.contains('SELL|STRONG_SELL',regex=True)]
```
